Visualizing the decision boundary of a kernelized SVM can be a bit more challenging than visualizing a linear SVM, especially in high-dimensional spaces. However, there are ways to project the data and decision boundary onto a 2D or 3D space for visualization purposes.

Here's an example of how to visualize the decision boundary of a kernelized SVM in Python using scikit-learn and matplotlib:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.svm import SVC

# Generate a non-linearly separable dataset
X, y = make_circles(n_samples=100, noise=0.1, random_state=1)

# Train a kernelized SVM with a radial basis function (RBF) kernel
clf = SVC(kernel='rbf', C=1, random_state=1)
clf.fit(X, y)

# Create a meshgrid to plot the decision boundary
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))

# Flatten the meshgrid and classify each point
X_grid = np.c_[xx.ravel(), yy.ravel()]
y_grid = clf.predict(X_grid)

# Plot the decision boundary
plt.contourf(xx, yy, y_grid.reshape(xx.shape), cmap=plt.cm.Paired)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')
plt.show()
```

In this example, we use the make_circles function to generate a non-linearly separable dataset, which consists of two classes of points arranged in the shape of two interleaved circles. Then we use the radial basis function (RBF) kernel to classify the points. The RBF kernel is a function that maps the input data into a higher-dimensional space where the data becomes linearly separable. The kernel trick makes it possible to find a linear boundary in the higher-dimensional space without the need to compute the coordinates of the data in that space.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.svm import SVC

# Generate a non-linearly separable dataset
X, y = make_circles(n_samples=100, noise=0.1, random_state=1)

# Train a kernelized SVM with a radial basis function (RBF) kernel
clf = SVC(kernel='rbf', C=1, random_state=1)
clf.fit(X, y)

# Create a meshgrid to plot the decision boundary
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))

# Flatten the meshgrid and classify each point
X_grid = np.c_[xx.ravel(), yy.ravel()]
y_grid = clf.predict(X_grid)

# Plot the decision boundary
plt.contourf(xx, yy, y_grid.reshape(xx.shape), cmap=plt.cm.Paired)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')
plt.show()
```
