In this example, we use the make_blobs function to generate a synthetic dataset with two classes of points that are linearly separable. Then, we use the SVC class from scikit-learn to train a maximum margin classifier (SVM) with a linear kernel. The C parameter is used to control the trade-off between maximizing the margin and minimizing the classification errors.

To visualize the decision boundary, we create a meshgrid that covers the entire range of the dataset, use the predict method of the trained SVM to classify each point in the grid, and plot the decision boundary using the contourf function from matplotlib.

You can try different datasets, kernels and parameters to see how the decision boundary changes and how it perform in different scenarios.

```python
from sklearn.datasets import make_blobs
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt

# Generate a synthetic dataset
X, y = make_blobs(n_samples=100, centers=2, random_state=1)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train a maximum margin classifier (SVM)
clf = SVC(C=1, kernel='linear')
clf.fit(X_train, y_train)

# Plot the decision boundary
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')
plt.show()
```
