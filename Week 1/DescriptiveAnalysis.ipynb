```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
```

Create a simple notebook and perform initial descriptive analysis (mean, mode, standard deviation, interquartile range, scatterplot)

```python
# Load dataset
data = pd.read_csv('my_dataset.csv')

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(data.head())

# Perform descriptive analysis
print("\nDescriptive Analysis:")
for column in data.columns:
    print(f"Column: {column}")
    print(f"Mean: {data[column].mean()}")
    print(f"Mode: {data[column].mode()[0]}")
    print(f"Standard Deviation: {data[column].std()}")
    print(f"Interquartile Range (IQR): {data[column].quantile(0.75) - data[column].quantile(0.25)}")
    print('\n')

# Scatter plot for the first two numeric columns
plt.figure(figsize=(10, 6))
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], alpha=0.7, color="blue")
plt.xlabel(data.columns[0])
plt.ylabel(data.columns[1])
plt.title(f"Scatter Plot: {data.columns[0]} vs {data.columns[1]}")
plt.grid(True)
plt.show()
```
